```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path="../Dataset/femaleHeaded.xlsx"
data=pd.read_excel(path)
```

```python
aggregated_data = data.groupby('Area Name')['Total number of female headed households'].sum().reset_index()
aggregated_data.rename(columns={'Area Name': 'region'}, inplace=True)
```

```python
aggregated_data['region'] = aggregated_data['region'].str.strip().str.lower()
```

```python
aggregated_data=aggregated_data[:-1]
```

```python
aggregated_data
```

```python
geojson_url = "https://raw.githubusercontent.com/Subhash9325/GeoJson-Data-of-Indian-States/refs/heads/master/Indian_States"
geo_data = gpd.read_file(geojson_url)
```

```python
print(geo_data['NAME_1'])
```

```python
geo_data['NAME_1'] = geo_data['NAME_1'].str.strip().str.lower()
aggregated_data['region'] = aggregated_data['region'].str.strip().str.lower()
```

```python
aggregated_data['region'] = aggregated_data['region'].replace('andaman & nicobar islands', 'andaman and nicobar')
aggregated_data['region'] = aggregated_data['region'].replace('dadra & nagar haveli', 'dadra and nagar haveli')
aggregated_data['region'] = aggregated_data['region'].replace('daman & diu', 'daman and diu')
aggregated_data['region'] = aggregated_data['region'].replace('jammu & kashmir', 'jammu and kashmir')
aggregated_data['region'] = aggregated_data['region'].replace('chattisgarh', 'chhattisgarh')
```

```python
aggregated_data['region'] = aggregated_data['region'].replace('nct of delhi', 'delhi')
aggregated_data['region'] = aggregated_data['region'].replace('uttarakhand', 'uttaranchal')
aggregated_data['region'] = aggregated_data['region'].replace('odisha', 'orissa')
```

```python
print("Normalized GeoJSON regions (NAME_1):", geo_data['NAME_1'].unique())
print("Normalized Dataset regions (region):", aggregated_data['region'].unique())
```

```python
merged = geo_data.merge(aggregated_data, left_on='NAME_1', right_on='region', how='left')  # 'st_nm' is the column for state names in this GeoJSON

# Plot the choropleth map
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged.boundary.plot(ax=ax, linewidth=1)
merged.plot(
    column='Total number of female headed households',  # Change to your column of interest
    cmap='OrRd',  # Choose a colormap
    legend=True,
    ax=ax
)

# Add title
plt.title('Choropleth Map of Female-Headed Households in Indian States', fontsize=16)
plt.axis('off')
plt.show()
```
